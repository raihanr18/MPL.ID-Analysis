```python
from sklearn.cluster import KMeans
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Datasets/MPL_ID_S10.csv')

df
```

```python
df = df.replace(to_replace = '-',value = np.nan)

df[["T_winrate", "T_pickpercentage","Ban_percentage"]] = df[[
    "T_winrate", "T_pickpercentage","Ban_percentage"]].apply(lambda x: x.str.replace("%", "").astype(float)/100)
```

```python
def dataframe_check(df):
    print('SHAPE'.center(82, '='))
    print(f'Rows: {df.shape[0]}')
    print(f'Column: {df.shape[1]}')
    print('TYPES'.center(82, '='))
    print(df.dtypes)
    print('CHECK THE AMOUNT OF MISSING DATA'.center(82, '='))
    print(f'Missing data: \n{df.isnull().sum()}')
    print('CHECK THE NUMBER OF INVALID DATA'.center(82, '='))
    print(f'Invalid data: \n{df.isna().sum()}')
    
dataframe_check(df)
```

```python
df = df.dropna()

df
```

```python
X = df[['T_winrate', 'T_pickpercentage', 'Ban_percentage']]

kmeans = KMeans(n_clusters=3)
y_kmeans = kmeans.fit_predict(X)

df = df.assign(cluster=y_kmeans)

df
```
